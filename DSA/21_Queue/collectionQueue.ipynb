```python
from collections import deque

customQueue=deque(maxlen=3)
customQueue.append(1)
customQueue.append(2)
customQueue.append(3)
customQueue.append(4) # it will remove 1 as maxlen is 3
print(customQueue)
```

```python
customQueue.popleft()
print(customQueue)
```

```python
## predefined functions for stack and queue with example 
from collections import deque
customQueue=deque()
customQueue.append(1)
customQueue.append(2)
customQueue.append(3)
print(customQueue)
customQueue.popleft()
print(customQueue)

## stack  by predefined library function latest 
from collections import deque
customStack=deque()
customStack.append(1)
customStack.append(2)
customStack.append(3)
print(customStack)
customStack.pop()
print(customStack)

```
