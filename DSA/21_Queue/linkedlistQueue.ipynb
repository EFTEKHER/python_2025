```python
class Node:
    def __init__(self,value):
        self.value=value
        self.next=None
    def __str__(self):
        return self.value    
        

class Queue:
    def __init__(self):
        self.length=0
        self.tail=None
        self.head=None
        
    def isEmpty(self):
        if self.head is None:
            return True
        else:
            return False
          
    def enqueue(self,value):
        node=Node(value)
        if self.isEmpty():
            self.head=node
            self.tail=node
            node.next= self.tail
        else:
            self.tail.next=node
            self.tail=node
    def __str__(self):
        if self.head is None:
            return "Empty"
        temp= self.head
        s=""
        while temp != self.tail:
            x = temp.value
            s += str(x) + " "
            temp = temp.next
        s+=str(self.tail.value)    
        return s
    def dequeue(self):
        if self.isEmpty():
            return "it is empty"
        else:
            self.head=self.head.next
            if self.head == self.tail:
                self.head=None
                self.tail=None
                
            
                    
            
        
        
        
        
    
```
