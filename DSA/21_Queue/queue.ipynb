```python
class Queue1:
    def __init__(self, maxsize):
        self.maxsize = maxsize
        self.lists = [None] * maxsize
        self.start = -1   # index of head
        self.top = -1     # index of tail
        self.size = 0

    def isfull(self):
        return self.size == self.maxsize

    def isEmpty(self):
        return self.size == 0

    def enqueue(self, value):
        if self.isfull():
            return "out of bound"
        self.top = (self.top + 1) % self.maxsize
        self.lists[self.top] = value
        if self.start == -1:           # first element inserted
            self.start = self.top
        self.size += 1
        return "value inserted"

    def dequeue(self):
        if self.isEmpty():
            return 'the list is empty'
        x = self.lists[self.start]
        self.lists[self.start] = None
        if self.size == 1:
            # queue becomes empty
            self.start = -1
            self.top = -1
            self.size = 0
        else:
            self.start = (self.start + 1) % self.maxsize
            self.size -= 1
        return x

    def peek(self):
        if self.isEmpty():
            return 'the list is empty'
        return self.lists[self.start]

    def __str__(self):
        if self.isEmpty():
            return ''
        out = []
        idx = self.start
        for _ in range(self.size):
            out.append(str(self.lists[idx]))
            idx = (idx + 1) % self.maxsize
        return ' '.join(out)
```
