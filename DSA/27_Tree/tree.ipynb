```python
class Tree:
    def __init__(self,data,children=[]):
        self.data=data
        self.children=children
    def __str__(self, level=0):
        r= " "*level+str(self.data)+"\n"
        for child in self.children:
            r+=child.__str__(level+1)
        return r
    def addchild(self,Tree):
        self.children.append(Tree)
        
            
        
            
```

```python
drinks=Tree('Drinks',[])
hot=Tree('Hot',[])
cold=Tree('Cold',[])
drinks.addchild(hot)
drinks.addchild(cold)
print(drinks)
coffee=Tree('coffee',[])
Tea=Tree('Tea',[])
hot.addchild(coffee)
hot.addchild(Tea)
print(drinks)
cold.addchild(Tree('Coke',[]))
cold.addchild(Tree('Pepsi',[]))
print(drinks)
```
