```python
def factorial(n):
    assert n>0 and int(n)==n ,'cannot negative or non integer'
    if n in [0,1]:
        return 1
    else:
        return n*factorial(n-1)
    
```

```python
factorial(-1)
```

```python
def sumofDigit(n):
    if n==0:
        return 0
    return (n//10)*10+sumofDigit(n//10)
```

```python
print(sumofDigit(123))
```
