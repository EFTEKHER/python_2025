```python
import matplotlib.pyplot as plt
```

```python
x=[1, 2, 3, 4, 5]  
y=[5, 4, 3, 2, 1]
y1=[8, 9, 10, 11, 12]
# plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Plot')
```

```python
plt.plot(x,y,color='red',linestyle='--',marker='o')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
# plt.plot(x, y, color='red', linestyle='--', marker='o')

plt.subplot(2, 2, 1)
plt.plot(x, y, color='blue')
plt.title('Subplot 1')

plt.subplot(3, 4, 3)
plt.plot(x, y1, color='green')
plt.title('Subplot 2')
```

```python
## bar plot
plt.figure(figsize=(10, 5))
plt.bar(x, y, color='blue', label='Bar 1')
plt.bar(x, y1, color='green', label='Bar 2', bottom=y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Bar Plot')
plt.legend()
```

```python
## histogram
plt.figure(figsize=(10, 5))
plt.hist(y, color='blue', alpha=0.7, label='Histogram 1')
plt.hist(y1, color='green', alpha=0.7, label='Histogram 2', stacked=True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Histogram')
plt.legend()
```

```python
data ={
    'x': x,
    'y': y,
    'y1': y1
}
```

```python
plt.hist(data,bins=10)
```

```python
colors=["blue","green","orange","purple"]
plt.pie(data['y'], labels=data['x'], colors=colors, autopct='%1.1f%%')
```
